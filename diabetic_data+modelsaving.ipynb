```python
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_json
import numpy as np
import os
```

```python
dataset=np.loadtxt(r"C:\Users\archa\OneDrive\Desktop\Downloads\pima-indians-diabetes.csv",delimiter=',')
print(dataset)
```

```python
dataset.shape
```

```python
x=dataset[:,0:8]
y=dataset[:,8]
print(x)
```

```python
print(y)
```

```python
model=Sequential()
model.add(Dense(12,activation='relu',input_dim=8,kernel_initializer='uniform'))
model.add(Dense(8,activation='relu',kernel_initializer='uniform'))
model.add(Dense(1,activation='sigmoid',kernel_initializer='uniform'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=150,batch_size=10,verbose=1)
```

```python
scores=model.evaluate(x,y)
scores
```

```python
model.metrics_names
```

```python
model_jason=model.to_json()
with open('model.json','w') as json_file:
    json_file.write(model_jason)
```

```python
model.save_weights('model.weights.h5')
```

```python
json_file=open('model.json','r')
loaded_model=json_file.read()
json_file.close()
loaded_model1=model_from_json(loaded_model)
```

```python
model.save_weights('model.weights.h5')
print('model loaded successfully')
```

```python
pred=loaded_model1.predict(x)
pred
```

```python
loaded_model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
loaded_model1.evaluate(x,y)
```
